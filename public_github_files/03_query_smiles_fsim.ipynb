```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.figure import figaspect
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
df_list = [pd.read_pickle("03_pkl_files/each_canon_query_penicillin.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_nirmatrelvir.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_azt.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_lsd.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_fentanyl.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_acid_blue_25_free_acid.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_avobenzone.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_diphenylaminocarbazole.pkl")
            ]
names_list = [
    "penicillin",
    "nirmatrelvir",
    "azt",
    "lsd",
    "fentanyl",
    "acid_blue_25_free_acid",
    "avobenzone",
    "diphenylaminocarbazole"
]
all_smiles_list = [
    pen_smiles,
    nirmatrelvir_smiles,
    azt_smiles,
    lsd_smiles,
    fentanyl_smiles,
    acid_blue_25_free_acid_smiles,
    avobenzone_smiles,
    diphenylaminocarbazole_smiles,
]
for df, name, smiles_list in zip(df_list, names_list, all_smiles_list):
    print(f"\nName: {name}")
    df["features"] = df["features"].apply(lambda x: x/np.linalg.norm(x))
    arr = np.array(df['features'].tolist())
    sim = np.dot(arr,arr.T)
    data = sim
    fig, ax = plt.subplots()
    cmap = sns.color_palette("viridis", as_cmap=True)
    ax = sns.heatmap(data, 
                 vmin=0, 
                 vmax=1, 
                 cmap=cmap,
                 cbar_kws={'label': 'Cosine Similarity'},
                 xticklabels= ["RDKit Atom 0", "RDKit Atom n", "OEChem"],
                 yticklabels= ["RDKit Atom 0", "RDKit Atom n", "OEChem"],
                #  cbar=False,
                 annot=True,
                 linewidths=0.5,
                 square=True,
                 )
    ax.xaxis.tick_top() # x axis on top
    ax.tick_params(length=0)
    plt.setp(ax.get_yticklabels(), rotation=0, ha="right",
            rotation_mode="anchor")
    
    # Make the font size bigger on the annotation in the heatmap
    for text in ax.texts:
        text.set_size(18)
    cbar = ax.collections[0].colorbar
    cbar.ax.tick_params(labelsize=18)
    cbar.ax.set_ylabel(cbar.ax.get_ylabel(), fontsize=18)
    cbar.ax.yaxis.set_label_coords(5, 0.5)

    # Make the font size bigger on the axis labels
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.set_xlabel(ax.get_xlabel(), fontsize=12)
    ax.set_ylabel(ax.get_ylabel(), fontsize=12)
    ax.xaxis.set_label_coords(0.5, -0.05)
    ax.yaxis.set_label_coords(-0.05, 0.5)




    plt.savefig(f"query_canon_fsim_{name}.png", dpi=300, bbox_inches = "tight")
```
