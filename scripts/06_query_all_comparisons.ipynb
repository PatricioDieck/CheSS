```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.figure import figaspect
from difflib import SequenceMatcher
from transformers import AutoTokenizer
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

```python
def token_sim(smi1, smi2, tokenizer):
    smi1_tokens = tokenizer(smi1).input_ids
    smi2_tokens = tokenizer(smi2).input_ids
    ratio_token_len = len(smi1_tokens) / len(smi2_tokens)
    ratio_shared_tokens = len(set(smi1_tokens) & set(smi2_tokens)) / len(set(smi1_tokens) | set(smi2_tokens))
    return ratio_shared_tokens, ratio_token_len
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "seyonec/PubChem10M_SMILES_BPE_450k", model_max_length=512,
)



df_list = [pd.read_pickle("03_pkl_files/each_canon_query_penicillin.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_nirmatrelvir.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_azt.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_lsd.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_fentanyl.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_acid_blue_25_free_acid.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_avobenzone.pkl"),
            pd.read_pickle("03_pkl_files/each_canon_query_diphenylaminocarbazole.pkl")
            ]

names_list = [
    "penicillin",
    "nirmatrelvir",
    "azt",
    "lsd",
    "fentanyl",
    "acid_blue_25_free_acid",
    "avobenzone",
    "diphenylaminocarbazole"
]
all_smiles_list = [
    pen_smiles,
    nirmatrelvir_smiles,
    azt_smiles,
    lsd_smiles,
    fentanyl_smiles,
    acid_blue_25_free_acid_smiles,
    avobenzone_smiles,
    diphenylaminocarbazole_smiles,
]

fsim_agg = np.zeros((8,3,3))
gestalt_agg = np.zeros((8,3,3))
token_sim_agg = np.zeros((8,3,3))
token_len_ratio_agg = np.zeros((8,3,3))

for count, (df, name, smiles_list) in enumerate(zip(df_list, names_list, all_smiles_list)):
    df["features"] = df["features"].apply(lambda x: x/np.linalg.norm(x))
    arr = np.array(df['features'].tolist())
    fsim_agg[count] = np.dot(arr, arr.T)

for count_mol, (name, smiles_list) in enumerate(zip(names_list, all_smiles_list)):
    for count_x, smi_x in enumerate(smiles_list):
        for count_y, smi_y in enumerate(smiles_list):
            gestalt_agg[count_mol,count_x,count_y] = SequenceMatcher(None, smi_x, smi_y).ratio()
            token_sim_agg[count_mol,count_x,count_y], token_len_ratio_agg[count_mol,count_x,count_y] = token_sim(smi_x, smi_y, tokenizer)
```
